### dataset 2: Google Play Store Apps

we download this data from https://www.kaggle.com/lava18/google-play-store-apps

The Play Store apps data has enormous potential to drive app-making businesses to success. Some insights can be drawn for developers to work on and capture the Android market. 

There are 13 columns and 9660 rows in this dataset. Each row contains a unique App from Google Play Store 

#### Attributes of google play store apps:
* App: name of the App
* Category: type of the App
* Rating: application's rating on playstore
* Reviews: number of reviews
* Size: size of app in M, sometimes varies with device.
* Installs: number of instrallments
* Type: whether the app is *free* or *paid*
* Price: price of the App in dollar, if the app is free, price equals to zero
* Content Rating: everyone, teens, mature 17+ etc. 7 values total
* Genres: App genres(i.e. tools, education, lifestyle etc. 120 values total)
* Last Updated: data of latest update
* Current Ver: latest App version
* Android Ver: Minimum Android Version required to run the App

#### Problems:
* Why some apps are high-rated while other not?
* Is rating related to genres?
* Is high rating apps always updating?
* If we are developing a new app on Android, how to get our app high_rated?
* Given other information, can we predict the rating of an app?
